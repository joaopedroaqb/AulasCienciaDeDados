```python
X = [0, 1, 1, 2, 2, 2, 3, 3]
print('X: ', X)
```

```python
from statistics import mode
m = mode(X)
print('A moda de X: ', m)
```

```python
values = np.sort(np.unique(X))
freq = np.zeros(len(values))

ind = 0
for i in values:
  counter = 0
  for j in range(0,len(X)):
    if (X[j]==i):
      counter = counter+1
  freq[ind] = counter
  ind = ind + 1
print('X:', X)
print('Valores Possiveis: ', values)
print('Frenquencia dos valores: ', freq)

y_pos = np.arange(len(values))
plt.xticks(y_pos, values)
plt.bar(y_pos, freq)
plt.xlabel("Valores", fontsize = 20)
plt.ylabel("Frequencia", fontsize = 20)

plt.show()
```

```python
mu = 0
sigma = 10
Y = np.random.normal(mu,sigma, 500)
plt.hist(Y, density = True, bins = 50, lw = 0, alpha = .8)

m = np.mean(Y)
md = np.median(Y)
plt.vlines(m, 0, .05)
plt.vlines(md, 0, .05, linestyles = 'dashed')
plt.xticks(fontsize=14)
plt.yticks(fontsize = 14)
plt.xlabel("X", fontsize = 20)
plt.ylabel("P(x)", fontsize = 20)

plt.show()
```

```python
lbda = 100
beta = 1.0/lbda

Y = np.random.exponential(beta, 500)
plt.hist(Y, density = True, bins = 50, lw = 0, alpha = .8)

m = np.mean(Y)
md = np.median(Y)

plt.vlines(m, 0, lbda)
plt.vlines(md, 0, lbda, linestyles = 'dashed')
plt.xticks(fontsize=14)
plt.yticks(fontsize = 14)
plt.xlabel("X", fontsize = 20)
plt.ylabel("P(x)", fontsize = 20)

plt.show()
```
